<h2> 2. Variables </h2>

<p> A variable is a named container that stores data. You can think of it like a labeled box that holds a value.</p>

```python
# Number of countries in a region
num_countries = 5

# Name of a city
city_name = "Berlin"

# Average temperature
avg_temp = 23.5
```

```python
print(city_name)
```

<h2> 3. Basic Operations </h2>

<p> You can perform mathematical operations on numbers:</p>

```python
# Addition
total_area = 450.5 + 300.2
print("Total Area:", total_area)

# Subtraction
difference = 500 - 450.5
print("Difference:", difference)

# Multiplication
double_area = 450.5 * 2
print("Double Area:", double_area)

# Division
per_country_area = 450.5 / 5
print("Average Area per Country:", per_country_area)
```

<h2> 3. Lists </h2>

<p> A list is a collection of items (numbers, strings, etc.): </p>

```python
cities = ["Berlin", "Paris", "Madrid"]
print("Cities:", cities)
print("First city:", cities[0])
```

<h2> 4. Conditional Statements </h2>

<p> Python can make decisions using if:</p>

```python
temperature = 18

if temperature > 20:
    print("Warm weather")
else:
    print("Cool weather")
```

<h2> 5. Loops</h2> 

<p> Loops help to repeat tasks, useful for iterating over data: </p>

```python
# Print all cities
for city in cities:
    print("City:", city)
```

<h2> 6. Geo Data Example with Coordinates </h2>

<p> Python can store geographic coordinates as variables or lists: </p>

```python
# Latitude and Longitude of Berlin
berlin_coords = [52.52, 13.405]

# Multiple cities coordinates
city_coords = {
    "Berlin": [52.52, 13.405],
    "Paris": [48.8566, 2.3522],
    "Madrid": [40.4168, -3.7038]
}

# Print coordinates
for city, coords in city_coords.items():
    print(city, "is at", coords)
```

<h2> 7. Functions </h2>
<p> A function is a reusable block of code that performs a specific task.
Functions help you avoid repeating the same code and make your analysis easier to read.</p>

```python
def population_density(population, area_km2):
    density = population / area_km2
    return density
```

```python
population_density(120000,500)
```

<h2> 8. Simple Visualization – Histogram from a list </h2>

<p> A histogram shows how values are distributed.
For example, we can visualize how many times a city appears in a list.

First, we create a list of values and then visualize them. </p>

```python
cities = ['Islamabad' , 'Lahore' , 'Karachi' ,'Lahore' ] 
```

```python
import matplotlib.pyplot as plt

plt.hist(cities)
plt.xlabel("Cities")
plt.ylabel("Number of cities")
plt.title("Distribution of city populations")

plt.show()
```
