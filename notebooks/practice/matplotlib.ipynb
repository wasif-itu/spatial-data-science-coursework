```python
import matplotlib.pyplot as plt
import math
```

```python
x = range(1,40)
y = [(i ** 5) for i in x]
plt.plot(x, y)

x = range(1,40)
y = [(math.sqrt(i)) for i in x]

plt.plot(x, y)
plt.show()

```

```python
x = list(range(1,25))
y = [math.sin(i) for i in x]

plt.subplot(2, 1, 1)
plt.scatter(x, y,label=["sin","xval"])
plt.legend()
y = [math.cos(i) for i in x]
plt.subplot(2, 1, 2)
plt.xlim(0,20)
plt.ylim(-1,1)
plt.scatter(x, y)
plt.show()
```
